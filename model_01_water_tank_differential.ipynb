# Model 01: Water Tank Simulation

**Objective**: Simulate the behavior of a water tank using a first-order differential equation.

***Given:***
- Initial volume: 1000 liters
- Inflow: 200 L/hr (constant)
- Outflow: 500 L/hr from 6â€“10 AM, otherwise 100 L/hr

$$
\frac{dW}{dt} = I(t) - O(t)
$$

*Interpret the result over a 24-hour period.*


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
t = np.linspace(0, 24, 1000)
```

```python
def outflow(t):
    if 6 <= t%24 <= 10:
        return 500;
    else:
        return 100;
```

```python
def dWdt(W, t):
    I = 200
    O = outflow(t)
    return I - O
```

```python
W0 = 1000
W = odeint(dWdt, W0, t)

plt.plot(t, W)
plt.title('Water Level in Tank Over 24 Hours')
plt.xlabel('Time (hours)')
plt.ylabel('Water Level (L)')
plt.grid(True)
plt.show()
```
