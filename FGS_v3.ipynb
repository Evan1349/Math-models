```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[1, 4], [2, 5], [3, 6]])
mean = np.mean(X, axis=0)
cov = np.cov(X.T)
eigvals, eigvecs = np.linalg.eig(cov)

plt.scatter(X[:,0], X[:,1], label='Data points')
plt.axvline(mean[0], color='gray', linestyle='--')
plt.axhline(mean[1], color='gray', linestyle='--')
plt.scatter(mean[0], mean[1], color='red', label='Mean')


for i in range(2):
    vec = eigvecs[:, i]

    scale = 2 * np.sqrt(eigvals[i])
    plt.plot([mean[0], mean[0] + vec[0]*scale],
             [mean[1], mean[1] + vec[1]*scale],
             label=f"Eigvec {i+1}")

plt.title(f"Mean: {mean}, Cov:\n{cov}")
plt.legend()
plt.axis('equal')
plt.show()
```

This scatter cloud has its mean at $[2, 5]$, and the points are spread along the $x = y$ direction.

The covariance matrix’s eigenvectors align with this spread.

So the principal axis shows the shape’s strongest trend — and that’s exactly the idea behind PCA.
