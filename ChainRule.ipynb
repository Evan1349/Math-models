# Chain Rule Example

This notebook demonstrates how to apply the Chain Rule for differentiation and visualize its behavior numerically.

---

## Mathematical Statement

The Chain Rule states that for a composite function:
$$
\frac{d}{dx}[f(g(x))] = f'(g(x)) \cdot g'(x)
$$

**Example:**  
- $f(x) = \sin(x^2)$

---

## What we do in this notebook

1. **Symbolic Derivation:**  
   Use `sympy` to compute the derivative symbolically and confirm it matches the hand-calculated Chain Rule result:
   $$
   \frac{d}{dx} \sin(x^2) = \cos(x^2) \cdot 2x.
   $$

2. **Numerical Evaluation:**  
   Convert the symbolic expressions to numerical functions with `lambdify` and evaluate over a range of \( x \).

3. **Plot Original Function vs. Derivative:**  
   Use `matplotlib` to plot:
   - The original function $f(x) = \sin(x^2)$
   - Its derivative $f'(x)$

   on the same graph for visual comparison.

---

## Why we plot both curves

- The **original function curve** shows where the function increases and decreases, and where it has local maximum or minimum points.
- The **derivative curve** shows how the slope changes:
  - Where the derivative equals zero, the original function has a horizontal tangent (potential max/min).
  - The sign of the derivative indicates whether the function is increasing (positive) or decreasing (negative).

This visualization confirms that the symbolic Chain Rule result also matches the local behavior numerically.

---

## Key idea

This workflow shows that the Chain Rule is not just an abstract formula —  
it governs how local changes propagate through a composite function, and you can see this clearly by comparing both curves.

---

## Requirements

- `numpy`
- `sympy`
- `matplotlib`



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = sp.symbols('x')
f = sp.sin(x**2)
```

```python
dfdx = sp.diff(f, x)
```

```python
print("Function: ", f)
print("Symbolic derivative (Chain Rule): ", dfdx)
```

```python
f_func = sp.lambdify(x, f, 'numpy')
dfdx_func = sp.lambdify(x, dfdx, 'numpy')
```

```python
X = np.linspace(-2, 2, 200)
```

```python
Y = f_func(X)
Y_prime = dfdx_func(X)
```

```python
plt.plot(X, Y, label='f(x) = sin(x²)')
plt.plot(X, Y_prime, label="f'(x)")
plt.axhline(0, color='gray', lw=0.5)
plt.title("Chain Rule Example: Symbolic vs Numeric")
plt.xlabel("x")
plt.ylabel("Value")
plt.legend()
plt.grid(True)
plt.show()
```

## Interpretation: What to check after plotting

After plotting the original function and its derivative together, check the following points to confirm your Chain Rule is applied correctly:

---

### 1. Zero Crossings of the Derivative

- The points where the derivative curve crosses zero should match the local maximum or minimum points of the original function.
- This confirms the slope of the original function is zero at those points.

---

### 2. Sign of the Derivative

- Where the derivative is **positive**, the original function should be increasing.
- Where the derivative is **negative**, the original function should be decreasing.
- This shows that the local growth behavior matches the symbolic result.

---

### 3️. Smoothness & Continuity

- Both the original function and its derivative should be smooth and continuous if the function is differentiable everywhere.
- Any unexpected jumps or breaks may indicate an error in the symbolic expression or numeric evaluation.

---

## Final Check

If all these conditions hold:
- The symbolic Chain Rule result is correct.
- The numerical values match the expected local behavior.
- The visualization clearly shows the connection between the function and its derivative.

---
