# Implicit Differentiation – Tangent Lines on Curves

---

## What is this?

This notebook demonstrates **Implicit Differentiation**  
and how to find the derivative $\frac{dy}{dx}$ when $y$ is not isolated.

---

## Core idea

- Some curves cannot be written as explicit functions $y = f(x)$.
- We treat $y$ as an implicit function of $x$ and differentiate both sides.
- Chain Rule is used whenever $y$ appears.
- The general strategy is:
  1. Differentiate both sides with respect to $x$.
  2. Solve the resulting expression for $\frac{dy}{dx}$.

---

## What we’ll do

1. Define an example implicit curve (e.g., $ x^2 + y^2 = 25 $ )  
2. Use **sympy** to compute $\frac{dy}{dx}$ symbolically  
3. Use **numpy + matplotlib** to plot the curve and its tangent lines  
4. Confirm that the slopes match the symbolic result

---

## Goal

> Understand how implicit differentiation works  
> and visualize how the derivative defines tangent lines on curves.


```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x, y = sp.symbols('x y')

eq = x**2 + y**2 - 25

dy_dx = -sp.diff(eq, x) / sp.diff(eq, y)

print("dy/dx =", dy_dx)
```

```python
r = 5
theta = np.linspace(0, 2*np.pi, 400)
x_circle = r * np.cos(theta)
y_circle = r * np.sin(theta)
```

```python
points = [
    (3, 4),  
    (4, 3),
    (-3, 4),  
    (-4, -3), 
    (3, -4)   
]


plt.figure(figsize=(8, 8))
plt.plot(x_circle, y_circle, label='x² + y² = 25')

for (x0, y0) in points:
    slope = -x0 / y0  

    x_tangent = np.linspace(x0 - 2, x0 + 2, 10)
    y_tangent = slope * (x_tangent - x0) + y0
    plt.plot(x_tangent, y_tangent, '--', label=f'Tangent at ({x0},{y0})')

    plt.plot(x0, y0, 'ro')

plt.gca().set_aspect('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Circle with Tangent Lines (Implicit Differentiation)')
plt.legend()
plt.grid(True)
plt.show()
```

## Interpretation

Why do we draw multiple tangent lines for an implicit curve like $ x^2 + y^2 = 25 $?

- This circle is an example of an **implicit curve** that cannot be written as a single explicit function $ y = f(x) $. For each $ x $, there are generally two possible $ y $-values, so the curve is not single-valued.
- When we use **implicit differentiation**, we get a formula for the slope:
  $
  \frac{dy}{dx} = -\frac{x}{y}.
  $
  This slope depends on the specific point \((x, y)\) on the curve.

- Drawing tangent lines at several points helps us see:
  1. The slope formula applies **locally** at each point. It changes sign and value depending on which quadrant the point is in.
  2. The direction of the tangent line reflects the relationship between $ x $ and $ y $ through the negative ratio $-x/y$.  
  For example, in Quadrant I ($ x > 0, y > 0 $), the slope is negative; in Quadrant II $( x < 0, y > 0 )$, it becomes positive.
  3. Since the circle is symmetric, the tangent behavior should also be symmetric — multiple tangent lines make this clear.

> In short, plotting multiple tangent lines verifies that **implicit differentiation** correctly captures how the slope behaves around the entire curve, not just at a single point.
