**Objective:**  
Simulate how an object cools down in an ambient environment based on the first-order differential equation:

$$
\frac{dT}{dt} = -k (T - T_a)
$$

**Key Parameters:**
- Initial temperature \( T_0 \)
- Ambient temperature \( T_a \)
- Cooling constant \( k \)

**Tools:**  
`numpy`, `scipy.integrate.odeint`, `matplotlib`

---

üìò This notebook demonstrates the relationship between the rate of cooling and the temperature difference.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def dTdt(T, t, k, Ta):
    return -k*(T- Ta)
```

```python
cooling_cases=[
    {"T0":90, "Ta":25, "k":0.1},
    {"T0":90, "Ta":25, "k":0.2},
    {"T0":90, "Ta":25, "k":0.3}
]
```

```python
def cooling_model(T0, Ta, k, t):
    return odeint(lambda T, t: dTdt(T, t, k, Ta), T0, t).flatten()
```

```python
t = np.linspace(0, 60, 300)
```

```python
plt.figure(figsize=(8, 5))

for case in cooling_cases:
    T0 = case['T0']
    Ta = case['Ta']
    k = case['k']
    
    T = cooling_model(T0, Ta, k, t)
    plt.plot(t, T, label=f"k={k}")

plt.axhline(y=cooling_cases[0]['Ta'], color='gray', linestyle='--', label='Ambient Temp')
plt.title("Newton's Law of Cooling ‚Äì Multiple Scenarios")
plt.xlabel("Time (minutes)")
plt.ylabel("Temperature (¬∞C)")
plt.legend()
plt.grid(True)
plt.show()
```
