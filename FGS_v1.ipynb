## What we’ll do

2. Use numpy to confirm & visualize.

```python
import numpy as np

A = np.array([[2, 1],
              [1, 2]])

# Determinant
det_A = np.linalg.det(A)
print("Determinant:", det_A)

# Eigenvalues and eigenvectors
eigvals, eigvecs = np.linalg.eig(A)
print("Eigenvalues:", eigvals)
print("Eigenvectors:\n", eigvecs)
```

```python
import matplotlib.pyplot as plt

# Tiny dataset
X = np.array([[1, 4],
              [2, 5],
              [3, 6]])

mean = np.mean(X, axis=0)
cov = np.cov(X.T)

# Scatter plot
plt.scatter(X[:,0], X[:,1])
plt.axvline(mean[0], color='gray', linestyle='--')
plt.axhline(mean[1], color='gray', linestyle='--')

plt.title(f"Mean: {mean}, Cov:\n{cov}")
plt.show()
```

## Micro Pitch Prompt

Explain in your own words:
1. What does the shape of this cloud tell you?
2. How does covariance relate to spread?
3. If you see a tail to the right, what would you do to prepare for supply?
